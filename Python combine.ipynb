```python
df1 = pd.read_csv('inv.csv', dtype={'storenum':int})
df1['storenum'] = df1['storenum'].astype(int)
df1.head()
```

```python
df2 = pd.read_csv('avg_wait_combined.csv')
df2['store_number_ship'] = pd.to_numeric(df2['store_number_ship'], errors='coerce')
df2 = df2.dropna(subset=['store_number_ship'])
df2['store_number_ship'] = df2['store_number_ship'].astype('int')

df2 = df2.dropna(subset=['store_number_receive'])
df2['store_number_receive'] = df2['store_number_receive'].astype('int')
df2.head()
```

```python
df_m = pd.merge(df2, df1, how = 'inner', left_on = 'store_number_ship', right_on = 'storenum')
df_m = df_m[df_m['smaxi']>0.0]
df_m.head()
```

```python
df_m = df_m.loc[df_m.groupby(['store_number_receive', 'skunum']).weighted_avg_delivery_time.idxmin()].reset_index(drop=True)
```
